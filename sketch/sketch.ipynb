**Importing necessary modules**

```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
```

**Setting figure size**

```python
matplotlib.rcParams['figure.figsize']=[20,10]
```

**Reading The image file**

```python
img = cv2.imread('img.jpg')
```

**Displaying the loaded image**

```python
plt.imshow(img[:,:,::-1]) #since Matplotlib uses RGB whereas cv2 uses BGR
```

**Grayscaling**

```python
grayscaled_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayscaled_img, cmap = 'gray') #no bgr or rgb since its grayscale. cmap is used to plot is as such in gray.
```

**Blurring the grayscaled image**

```python
blurred_grayscaled_img = cv2.GaussianBlur(grayscaled_img, (101,101), 0) #arguments obtained by trail and error
```

```python
plt.imshow(blurred_grayscaled_img, cmap = 'gray')
```

**Dividing using cv2 and presenting Final sketch**

```python
sketch = cv2.divide(grayscaled_img, blurred_grayscaled_img, scale = 256.0)
```

```python
plt.imshow(sketch, cmap = 'gray')
```
